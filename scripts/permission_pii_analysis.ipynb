```python
import pandas as pd

# Filter to only relevant categories (bluetooth, location, wifi, audio)
permission_categories = ['bluetooth', 'location', 'wifi']

# Total number of unique packages across the dataset
total_packages = df['package_name'].drop_duplicates().count()

# Function to calculate overlaps for a given permission category
def compute_overlap_percentage(df, category):
    cat_df = df[df['permission_category'] == category]
    
    # Build sets of package names per finding type
    first_party = set(cat_df[cat_df['finding_type'] == 'first_party_findings']['package_name'])
    third_party = set(cat_df[cat_df['finding_type'] == 'third_party_findings']['package_name'])
    bt_package = set(cat_df[cat_df['finding_type'] == 'bt_package_findings']['package_name'])

    # Calculate disjoint and overlap counts
    only_first_party = len(first_party - third_party - bt_package)
    only_third_party = len(third_party - first_party - bt_package)
    only_bt_package = len(bt_package - first_party - third_party) if category == 'bluetooth' else 0

    first_and_third = len(first_party & third_party)
    first_and_bt = len(first_party & bt_package) if category == 'bluetooth' else 0
    third_and_bt = len(third_party & bt_package) if category == 'bluetooth' else 0

    all_three = len(first_party & third_party & bt_package) if category == 'bluetooth' else 0

    # Total unique packages for this category
    total_unique = len(first_party | third_party | bt_package)

    # Calculate percentages
    def percent(val):
        return val

    result = {
        'permission_category': category,
        'only_first_party': percent(only_first_party),
        'only_third_party': percent(only_third_party),
        'only_bt_package': percent(only_bt_package),
        'first_and_third': percent(first_and_third),
        'first_and_bt': percent(first_and_bt),
        'third_and_bt': percent(third_and_bt),
        'all_three': percent(all_three),
        'total_unique': percent(total_unique)  # Total % of apps affected in this category
    }

    return result

# Run for all categories
results = []
for category in permission_categories:
    results.append(compute_overlap_percentage(df, category))

# Convert to DataFrame
overlap_percentage_df = pd.DataFrame(results)
overlap_percentage_df
```
